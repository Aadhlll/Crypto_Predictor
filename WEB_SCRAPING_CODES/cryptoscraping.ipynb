```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Set up Selenium
driver = webdriver.Chrome()

# Define the URL
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
TOP_CRYPTO = set()  # Initialize an empty set to store unique symbols

for year in years:
    if year == 2024:
        url = f"https://coinmarketcap.com/historical/20240310/"
    else:
        url = f"https://coinmarketcap.com/historical/{year}1231/"
    
    # Open the URL in the browser
    driver.get(url)
    # Set the browser window size to less than 50 pixels
    driver.set_window_size(300, 1000)  # Set the window width to 800 pixels and height to less than 50 pixels
    # Function to scroll slowly to halfway down the page
    def scroll_halfway_down(driver):
        # Get current scroll height
        last_height = driver.execute_script("return document.body.scrollHeight")
        time.sleep(6)
        # Scroll halfway down slowly
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 10);")
        time.sleep(6)  # Adjust sleep time as needed
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 5);")
        time.sleep(6)  # Adjust sleep time as needed
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 4);")
        time.sleep(6) 

    # Scroll slowly to halfway down the page
    scroll_halfway_down(driver)

    # Parse the HTML content
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Find all elements with class="cmc-table-row"
    cmc_rows = soup.find_all("tr", class_="cmc-table-row")[:50]  # Limit to first 50 rows

    # Initialize an empty list to store symbols
    symbols = []

    # Iterate through each row
    for row in cmc_rows:
        # Find the symbol element within the row
        symbol_element = row.find("a", class_="cmc-table__column-name--name cmc-link")
        
        # Extract the text content if the symbol element is found
        if symbol_element:
            symbol_text = symbol_element.text.strip()
            symbols.append(symbol_text)
    TOP_CRYPTO.update(symbols)
    # Print the collected symbols
    print(year,":",symbols)

# Close the browser
driver.quit()

# Print the unique symbols
print(TOP_CRYPTO)
```

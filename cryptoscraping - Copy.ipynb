```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Set up Selenium
driver = webdriver.Chrome()

# Define the URL
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
TOP_CRYPTO = set()  # Initialize an empty set to store unique symbols

for year in years:
    if year == 2024:
        url = f"https://coinmarketcap.com/historical/20240310/"
    else:
        url = f"https://coinmarketcap.com/historical/{year}1231/"
    
    # Open the URL in the browser
    driver.get(url)
    # Set the browser window size to less than 50 pixels
    driver.set_window_size(300, 1000)  # Set the window width to 800 pixels and height to less than 50 pixels
    # Function to scroll slowly to halfway down the page
    def scroll_halfway_down(driver):
        # Get current scroll height
        last_height = driver.execute_script("return document.body.scrollHeight")
        time.sleep(6)
        # Scroll halfway down slowly
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 10);")
        time.sleep(6)  # Adjust sleep time as needed
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 5);")
        time.sleep(6)  # Adjust sleep time as needed
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 4);")
        time.sleep(6) 

    # Scroll slowly to halfway down the page
    scroll_halfway_down(driver)

    # Parse the HTML content
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Find all elements with class="cmc-table-row"
    cmc_rows = soup.find_all("tr", class_="cmc-table-row")[:50]  # Limit to first 50 rows

    # Initialize an empty list to store symbols
    symbols = []

    # Iterate through each row
    for row in cmc_rows:
        # Find the symbol element within the row
        symbol_element = row.find("a", class_="cmc-table__column-name--name cmc-link")
        
        # Extract the text content if the symbol element is found
        if symbol_element:
            symbol_text = symbol_element.text.strip()
            symbols.append(symbol_text)
    TOP_CRYPTO.update(symbols)
    # Print the collected symbols
    print(year,":",symbols)

# Close the browser
driver.quit()

# Print the unique symbols
print(TOP_CRYPTO)
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import time

# Set up the WebDriver

driver = webdriver.Chrome()
tickers = ['%5EGSPC', '%5EIXIC', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'PYPL', 'SQ', 'V', 'MA', 'COIN', 'MSTR', 'RIOT', 'MARA', 'GC=F', 'SI=F']

# Initialize an empty list to store DataFrames
all_data = []

# Loop through each ticker
for ticker in tickers:
    url = f'https://finance.yahoo.com/quote/{ticker}/history/?period1=1420070400&period2=1718101333'
    try:
        driver.get(url)
        # Optionally wait for the page to fully load
        time.sleep(10)

        # Get the page source
        html = driver.page_source

        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')

       
        # Find the table container
        table_container = soup.find('div', class_='table-container')

        if table_container:
            # Find all rows in the table
            rows = table_container.find_all('tr', class_='svelte-ewueuo')

            # Extract header data
            headers = [th.text.strip() for th in rows[0].find_all('th')]

            # Extract row data
            data = []
            for row in rows[1:]:
                cells = row.find_all('td')
                row_data = [cell.text.strip() for cell in cells]
                if row_data:  # Only add rows with data
                    data.append(row_data)

            # Create DataFrame
            df = pd.DataFrame(data, columns=headers)
            df['Ticker'] = ticker  # Add a column for the ticker symbol

            # Append the DataFrame to the list
            all_data.append(df)
        else:
            print(f'Table container not found for {ticker}')
    except Exception as e:
        print(f"URL not working for {ticker}. Error: {e}")

# Close the WebDriver after the loop
driver.quit()

# Concatenate all DataFrames into one
final_df = pd.concat(all_data, ignore_index=True)

# Print the final DataFrame
print(final_df)
```

```python
import os
import pandas as pd
import numpy as np  # Import numpy for NaN values

# Path to the directory containing all the CSV files
directory = r'D:\DATA SCIENCE\Internship\CrypoDataAnalysis\COINMARKETDATA'

# List all CSV files in the directory
files = os.listdir(directory)

# Initialize an empty list to store DataFrames for each coin
coin_dfs = []

# Iterate through each CSV file
for file in files:
    if file.endswith('.csv'):
        # Read the CSV file
        df = pd.read_csv(os.path.join(directory, file))
        
        # Replace "No data is available now" with NaN
        df['Date'] = df['Date'].replace("No data is available now", np.nan)
        
        # Convert 'Date' column to Date format
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce', format='%b %d, %Y')  # Specify date format and handle errors
        
        # Drop rows with NaT (Not a Time) values resulting from conversion errors
        df = df.dropna(subset=['Date'])
        
        # Keep only 'Date' and 'Close' columns
        df = df[['Date', 'Close']]
        
        # Rename 'Close' column to the coin name
        coin_name = file.split('_')[0]  # Extract coin name from file name
        df.rename(columns={'Close': coin_name}, inplace=True)
        
        # Append DataFrame to the list
        coin_dfs.append(df)

# Merge all DataFrames on the 'Date' column
combined_df = coin_dfs[0]
for df in coin_dfs[1:]:
    combined_df = pd.merge(combined_df, df, on='Date', how='outer')

# Sort DataFrame based on 'Date'
combined_df.sort_values(by='Date', inplace=True)

# Reset index
combined_df.reset_index(drop=True, inplace=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('combinedd_coins_data.csv', index=False)

# Print the combined DataFrame
print(combined_df)
```

```python
import pandas as pd
import os

# Directory containing all CSV files
directory = 'D:\\DATA SCIENCE\\Internship\\CrypoDataAnalysis'

# List all CSV files in the directory
csv_files = [file for file in os.listdir(directory) if file.endswith('historical_data.csv')]

# Initialize an empty DataFrame to store combined data
combined_df = pd.DataFrame()
bitcoin_df = pd.DataFrame()  # DataFrame to store Bitcoin data

# Iterate over each CSV file
for file in csv_files:
    # Read the CSV file
    df = pd.read_csv(os.path.join(directory, file))
    # Convert 'Date' column to Date format
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce', format='%b %d, %Y')  # Specify date format and handle errors
    # Ensure 'Volume' is treated as string before cleaning and converting
    df['Volume'] = df['Volume'].astype(str).str.replace('[$,]', '', regex=True).astype(float)
    df['Market Cap'] = df['Market Cap'].astype(str).str.replace('[$,]', '', regex=True).astype(float)
    # Extract coin name from the file name
    coin_name = file.split('_')[0]
    
    if coin_name.lower() == 'bitcoin':
        # For Bitcoin, consider all columns and set 'Date' as index
        df.set_index('Date', inplace=True)
        bitcoin_df = df  # Store Bitcoin data separately
    else:
        # For other coins, consider only 'Close' and 'Volume' columns
        df = df[['Date', 'Close', 'Volume']].set_index('Date')
        # Rename the 'Close' and 'Volume' columns with coin name
        df.rename(columns={'Close': coin_name, 'Volume': f'{coin_name}_volume'}, inplace=True)
        # Concatenate with the combined DataFrame
        combined_df = pd.concat([combined_df, df], axis=1, sort=False)

# Calculate the maximum volume for each coin (excluding Bitcoin)
volumes = combined_df.filter(like='_volume').max()

# Sort coins by their maximum volume in descending order
sorted_volumes = volumes.sort_values(ascending=False)

# Get the sorted list of column names
sorted_columns = []
for coin in sorted_volumes.index:
    coin_name = coin.replace('_volume', '')
    sorted_columns.extend([coin_name, coin])

# Concatenate Bitcoin data with the combined DataFrame and sorted columns
combined_df = pd.concat([bitcoin_df, combined_df[sorted_columns]], axis=1)
combined_df = combined_df[combined_df.index.notnull()]
combined_df = combined_df.drop(pd.Timestamp('2024-05-02'))
pd.set_option('display.max_columns', None) 
# Columns to be deleted
columns_to_delete = [
    'bitcoindark', 'bitcoindark_volume', 'bitconnect', 'bitconnect_volume',
    'factom', 'factom_volume', 'iconomi', 'iconomi_volume',
    'insight-chain', 'insight-chain_volume', 'nubits', 'nubits_volume',
    'thorenext', 'thorenext_volume'
]

# Drop the specified columns
combined_df.drop(columns=columns_to_delete, inplace=True, errors='ignore')

# Display the DataFrame after removal
combined_df.tail(50)

```
